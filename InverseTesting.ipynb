```
from numpy import loadtxt
from keras.models import Sequential #Sequential helps us add each layer of the NN step by step#
from keras.layers import Dense #A Dense layer feeds all outputs from the previous layer to all its neurons, each neuron providing one output to the next layer. It's the most basic layer in neural networks. A Dense(10) has ten neurons.#
from keras import optimizers
# load the dataset
from google.colab import files
uploaded = files.upload()
dataset = loadtxt('ForwardTrain.csv', delimiter=',')
test = loadtxt('ForwardTest.csv', delimiter=',')
# split into input (X) and output (y) variables using slice operator
X = dataset[:,0:1] #taking all rows (:) but keeping the second column (1)
y = dataset[:,1:12]
X_pred = test[:,0:1]

# define the keras model
model = Sequential()
model.add(Dense(256, input_dim=1, activation='relu'))    #256 is the number of neurons#
model.add(Dense(128, activation='relu'))
model.add(Dense(11))

# compile the keras model
# sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='mean_squared_error', optimizer='adam')

# fit the keras model on the dataset
model.fit(X, y, epochs=5000, batch_size=3, verbose=0) #The batch size is a number of samples processed before the model is updated#
																												#number of epochs is the number of complete passes through the training dataset. #
# make class predictions with the model
predictions = model.predict(X_pred)
# summarize the first 2 cases
for i in range(2):
	print(predictions[i])
 
# # evaluate the keras model
# _, accuracy = model.evaluate(X, y, verbose=0)
# print('Accuracy: %.2f' % (accuracy*100))
# predictions = model.predict_classes(X)
# # summarize the first 5 cases
# for i in range(5):
# 	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
